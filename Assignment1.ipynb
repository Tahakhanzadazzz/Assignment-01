Function for web scraping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape data from the e-commerce site
def scrape_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract title
    title_element = soup.find('title')
    title = title_element.text.strip() if title_element else None

    # Extract price
    price_element = soup.find('span', class_='a-offscreen')
    price = float(price_element.text.strip('$')) if price_element else None

    # Create a dictionary to store the scraped data
    data = {'Title': title, 'Price': price}
    return data

# URL of the product page
url = "https://www.amazon.com/SAMSUNG-Factory-Unlocked-Android-Smartphone/dp/B0BLP2G96N"

# Scrape data from the product page
data = scrape_data(url)

# Check if data is retrieved successfully
if data is not None:
    # Create a Pandas DataFrame
    df = pd.DataFrame([data])

    # Display the scraped data
    print(df)
```

# Function for cleaning

```python
def clean_data(df):
    # Check for duplicates and missing values
    cleaned_df = df.drop_duplicates()  # Remove duplicates
    cleaned_df = cleaned_df.dropna()   # Remove rows with missing values
    return cleaned_df
```

```python
# URLs of the two e-commerce sites
url1 = "https://www.amazon.com/SAMSUNG-Factory-Unlocked-Android-Smartphone/dp/B0BLP2G96N/ref=sr_1_1_sspa?crid=22ILDX7TORKMC&keywords=Mobile&qid=1699869286&sprefix=mobi%2Caps%2C384&sr=8-1-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&th=1"
url2 = "https://www.daraz.com/"
```

```python

# Scrape data from the two e-commerce sites
data1 = scrape_data(url1)
data2 = scrape_data(url2)

# Create Pandas DataFrames
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Clean the data
cleaned_df1 = clean_data(df1)
cleaned_df2 = clean_data(df2)

# Compare prices and recommend the e-commerce site
average_price1 = cleaned_df1['Price'].mean()
average_price2 = cleaned_df2['Price'].mean()


# Display the scraped and cleaned data
print("Scraped Data from", url1)
print(df1)
print("\nCleaned Data from", url1)
print(cleaned_df1)

print("\nScraped Data from", url2)
print(df2)
print("\nCleaned Data from", url2)
print(cleaned_df2)

if average_price1 < average_price2:
    print("Recommendation: Choose the product from the first e-commerce site.")
else:
    print("Recommendation: Choose the product from the second e-commerce site.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to scrape data from the e-commerce site
def scrape_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Example: Extract product name and price
    product_name_element = soup.find('h2', class_='productTitle')
    price_element = soup.find('span', class_='a-price-whole')

    # Check if elements are found
    if product_name_element and price_element:
        product_name = product_name_element.text.strip()
        price = float(price_element.text.strip('$'))

        # Create a dictionary to store the scraped data
        data = {'Product Name': product_name, 'Price': price}
        return data
    else:
        print("Product information not found on the page.")
        return None

# URL of the product page
url = "https://www.amazon.com/SAMSUNG-Factory-Unlocked-Android-Smartphone/dp/B0BLP2G96N/ref=sr_1_1_sspa?crid=22ILDX7TORKMC&keywords=Mobile&qid=1699869286&sprefix=mobi%2Caps%2C384&sr=8-1-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&th=1"

# Scrape data from the product page
data = scrape_data(url)

# Check if data is retrieved successfully
if data is not None:
    # Create a Pandas DataFrame
    df = pd.DataFrame([data])

    # Display the scraped data
    print(df)
```
